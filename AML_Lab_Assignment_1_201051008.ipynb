# 3. Rewrite the program-1 using try and except to prompt the user for hours and rate per hour to  compute gross pay. (if non-numeric inputs entered, it should except). Hour = 35 & Rate = 2.75.

```
def grossPay(hours, rate):
    return hours*rate

while(True):
    try:
        hours = float(input('Enter the number of hours : '))
        break
    except ValueError:
        print("Invalid value entered please enter a numeric value.")

while(True):
    try:
        rate = float(input('Enter the rate : '))
        break
    except ValueError:
        print("Invalid value entered please enter a numeric value.")

gross_pay = grossPay(hours, rate)
print("Gross pay is ", gross_pay)
```

# 4. Rewrite the program-1 with time-and-a half for overtime and create a function called paycomp which takes two parameters (hours and rate per hour). Hours = 45 and rate = 10.

```
def paycomp(hours, rate):
    overtime_limit = 30
    if hours <= overtime_limit:
        return hours*rate
    else:
        return (overtime_limit*rate) + ((hours-overtime_limit)*rate*1.5)
        
print("grosspay is ", paycomp(45, 10))
```

# 7. Consider the list of elements [9, 41, 23, 54, 33, 21, 8] use for loop to find
a. Largest number 

b. Smallest number

c. Number of numbers

d. Number of odd number

e. Number of even numbers

f. Number of prime numbers

g. Sum and average of numbers

h. Filter the numbers greater than 20

i. Filter the numbers less than 15

j. Search for the number 3

```
listt = [9, 41, 23, 54, 33, 21, 8] 
```

```
# a. Largest number
largest_number = listt[0]
for element in listt:
    if element > largest_number:
        largest_number = element

print("Largest number is ", largest_number)
```

```
# b. Smallest number
smallest_number = listt[0]
for element in listt:
    if element < smallest_number:
        smallest_number = element

print("Smallest number is ", smallest_number)
```

```
# c. Number of numbers
count = 0
for element in listt:
    count = count + 1

print("The total number of elements in the list is ", count)
```

```
# d. Number of odd number
count = 0
for element in listt:
    if element % 2 == 1:
        count = count + 1
print("The number of odd numbers in the list is ", count)
```

```
# e. Number of even numbers
count = 0
for element in listt:
    if element % 2 == 0:
        count = count + 1
print("The number of even numbers in the list is ", count)
```

```
# f. Number of prime numbers
def isprime(num):
    flag = True
    for i in range(2, num):
        if num % i == 0:
            flag = False
            break
    return flag 

count = 0
for element in listt:
    if isprime(element):
        count = count + 1
print("The number of prime numbers in the list is ", count)
```

```
# g. Sum and average of numbers

def sum(list_num):
    total = 0
    for i in list_num:
        total  = total + i
    return total

def avg(list_avg):
    avg_list = sum(list_avg) / len(list_avg)
    return avg_list

print("The sum of numbers in the list is ", sum(listt))
print("The average of numbers in the list is ", avg(listt))
```

```
# h. Filter the numbers greater than 20

new_list = listt[:]
for element in listt:
    if element > 20:
        new_list.remove(element)

print("The filtered list of numbers is ",new_list)
 
```

```
# i. Filter the numbers less than 15

new_list = listt[:]
for element in listt:
    if element < 15:
        new_list.remove(element)

print("The filtered list of numbers is ",new_list)
 
```

```
# j. Search for the number 3

found = False
for i, element in enumerate(listt):
    if element == 3:
        print("3 Found in the list at index ", i)
        found = True
        break 
if found == False:
    print("Element not found in list")
```

# 8. Illustrate the use of type operator and type conversion (use your own examples)

## 9. Illustrate the use of break and continue with your own examples

```
eg_list = [ 1, 2, 3, 4, 5, 5, 4, 3, 2, 1 ]

print('Searching using for loop and using break' )

for i, num in enumerate(eg_list):
    if num == 3:
        print("found 3 and the index is ", i)
        break

print('\n')
print('Searching using for loop and using Continue' )

for i, num in enumerate(eg_list):
    if num == 3:
        print("found 3 and the index is ", i)
        continue

print('\n')
print('Break will stop the loop once found the element \nbut continue will continue the loop, so then we will find all the index ' )
```
